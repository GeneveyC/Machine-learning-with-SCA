```python
import matplotlib.pyplot as plt
import numpy as np
```

## Training history
Ensure models were correctly trained

```python
archi_list = {
    'ascad': '#ff7f0e',
    'zaid': '#2ca02c',
    'noConv1': '#d62728',
}
metrics = ('loss','accuracy','mean_rank')
# archi / device / source / metric
prefix = "../models_gan/stm32/{}/transfer/1kfold_trace_BD_SNR4_1_08B{}{}/"
history_format = prefix + "hist_1.npy"
model_format   = prefix + "dnn_1_best-{}.hdf5"
```

```python
devices = ('f0','f1','f2','f3','f4')
sources = ('pw','em')
```

## Core function to plot evolution

```python
def plot_evolution(device,source):
    
    axes = []
    for i, metric in enumerate(metrics):
        fig = plt.figure()
        fig.set_figheight(6)
        fig.set_figwidth(12)
        axes.append(fig.gca())
    
    for archi, color in archi_list.items():
        history = np.load(history_format.format(archi,device.upper(),source.upper()),allow_pickle=True).item()
        for i, metric in enumerate(metrics):
            axes[i].plot(history[metric],color=color,ls='--',label='{} (train)'.format(archi))
            axes[i].plot(history['val_'+metric],color=color,label='{} (val)'.format(archi))
    for i, metric in enumerate(metrics):
        axes[i].set_title(metric)
        axes[i].legend()
        
    fig.show()
```

## F0 PW

```python
plot_evolution('f0','pw')
```
