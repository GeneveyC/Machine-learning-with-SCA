```python
def plot_and_save(file, corr_max, steper, limit=None, key_n=0):
    plt.figure()
    if limit==None:
        plt.plot(steper[:],corr_max[:,:], "gray");
        plt.plot(steper[:],corr_max[:,1], "gray", label = "wrong candidates");
        plt.plot(steper[:],corr_max[:,key_n], "k", label="correct key");
    else:
        plt.plot(steper[:limit],corr_max[:limit,:], "gray");
        plt.plot(steper[:limit],corr_max[:limit,1], "gray", label = "wrong candidates");
        plt.plot(steper[:limit],corr_max[:limit,key_n], "k", label="correct key");        
    plt.grid()
    plt.xlabel("Nr of traces on the target device", fontsize=18);
    plt.xlim([min(steper), max(steper)])
    tmp = np.asarray(steper)
    if tmp[-1] > 200:
        plt.xticks(range(100,1000,100))
    else:
        plt.xticks(tmp[::2])
    plt.ylabel("Correlation", fontsize=18);
    plt.legend(fontsize="x-large")
    plt.savefig(file[:-4] +".png")
    print("saved "+ file[:-4] +".png")
    
```
